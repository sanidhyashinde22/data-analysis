### <a id='toc1_1_1_'></a>[What is Exploratory Data Analysis](#toc0_) 

* EDA is an acronym for `Exploratory Data Analysis `. It is a method of data analysis that employs visual techniques. With the aid of statistical summaries and graphical representations, it is used to find trends, patterns, or to verify hypotheses.



* It assists in preparing data for analysis.
* It can yield more accurate findings.
* It also assists us in selecting a superior machine learning model.

### <a id='toc1_4_2_'></a>[Handling the Null values.](#toc0_)
 One can use various methods on different features depending on how and what the data is about.
 Data cleaning is the process of locating and rectifying any faulty or incorrect data in a dataset. One such method is to do something about the missing values in the dataset. Numerous datasets will include many missing values in real life, therefore dealing with them is a critical step.


* Filling with Median Vlaues using `median()` values.

```python
df['household_size'].fillna(df['household_size'].median(), inplace = True)
df['household_size']
```

* Filling values with Mean values.

* Here we are using `mean()` method.

```python
df['Respondent Age'].fillna(df['Respondent Age'].mean(), inplace = True)
df['Respondent Age']
```

* We have scikit learn imputer, but it works only for numerical data. So we have sklearn_pandas with the transformer equivalent to that, which can work with string data. It replaces missing values with the most frequent ones in that column. 

```python
imputer = SimpleImputer(strategy="most_frequent")
data = np.array(df['Type of Job'], dtype=object).reshape(-1, 1)
imputer.fit_transform(data)
```

```python
imputer = SimpleImputer(strategy="most_frequent")
data = np.array(df['Level of Educuation'], dtype=object).reshape(-1, 1)
imputer.fit_transform(data)
```

```python
imputer = SimpleImputer(strategy="most_frequent")
data = np.array(df['marital_status'], dtype=object).reshape(-1, 1)
imputer.fit_transform(data)
```

* One such method is to do something about the missing values in the dataset. Numerous datasets will include many missing values in real life, therefore dealing with them is a critical step.

* Here we are using `value_counts()`method.

```python
# filling with most common class
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df 
```

### <a id='toc1_4_5_'></a>[Countplot](#toc0_)
* One way to think of a count plot is as a histogram that is plotted over a categorical variable rather than a quantitative one.<br>
* Bars are used to show how many observations fit into each group of classification. 
<br>
You can read more about it on <a href="https://seaborn.pydata.org/generated/seaborn.countplot.html#:~:text=Show%20the%20counts%20of%20observations,compare%20counts%20across%20nested%20variables.">This link</a>.


```python
plt.tick_params(axis='both', which='major', labelsize=14) #it will increase the fontsize of the label
sns.countplot(x=df["gender_of_respondent"],palette='Paired') #here we are defining the values of x axis and color palette
```

As you can see there are many female customers as compaired to thee male customers

```python
plt.figure(figsize=(10,5))
plt.tick_params(axis='both', which='major', labelsize=14) #it will increase the fontsize of the label
plt.xticks(rotation=45)
sns.countplot(x=df["Type of Job"],palette='Paired') #here we are defining the values of x axis and color palette
```
